## 3. Methodology

### 3.1 Exploratory Data Analysis

Neighborhoods needed to be examine separetly in order to select the best location for opening a coffee shop. Foursquare API is used to explore venues in the neighbourhoods in Toronto. There are 38 neighborhoods in Toronto. Top 100 venues for each neighborhood are requested from Foursquare API. The resulting data includes the name and category as well as the latitude and longitude of the each venue. 

In the first part of the analysis, the resulting data filtered into the category of 'Coffee Shop'. The frequency of occurrence of coffee shops are calculated for each neighborhood. k-means clustering method is used to analyze the data. By this method, neighborhoods can be split into several clusters. The number of clusters, which is k, is determined to be 3 for this analysis. Clusters are visualized as a map, shown below.

## 4. Results

As explained in the previous section, k-means clustering method is applied into the dataset. 3 different clusters are obtained. Cluster 0 is the cluster of neighborhoods that have moderate number of coffee shops, while Cluster 1 is high and Cluster 2 is low. The goal of the project is determining the best location to open a coffee shop. If there is not many coffee shops in the area, the competition will be low there. So the best choice will be a neighborhood from Cluster 2. 

The best location is selected from the analysis up to this point, however there are 15 neighborhoods in the selected cluster. It is still a high number of neighborhoods cluster to choose from. In order to reduce the number, one more k-means cluster algorithm is run, but this time most common venues of neighborhoods are used. In this further analysis, the clusters are checked for whether their most common venues are coffee shops and cafes or not. If the coffee shop is the 1st most common venue of the neighborhood, it means there is a demand for a coffee shop there. According to the resulting clusters, cluster 0 have their 1st most common venues as Cafes, and Coffee Shops. The other venues are also important for our decision. In cluster 0, there are other most common venues such as hotel, park, theatre which can increase the customers of coffee shop. 

Consequently, the analysis shows that the following Neighborhoods are the most suitable ones to open a coffee shop:
